On considère la plaque chargée uniformément en volume avec une charge Q.

A présent calculons E.

On se place dans l'approximation de la plaque infinie en négligeant ainsi les effets de bord. Par symétrie, on a donc
$\vec{E} = E \vec{e_y}$  

---

### Application du Théorème de Gauss :

$$
\oint_{S_1} \vec{E} \cdot \mathrm{d}\vec{S_1} \, \vec{e_y} 
+ \oint_{S_2} (-\vec{E}) \cdot (-\mathrm{d}\vec{S_2}) \, \vec{e_y} 
+ \oint_{S_3} \vec{E} \cdot \mathrm{d}\vec{S_3} \, \vec{u} 
= \frac{Q_\text{int}}{\varepsilon_0}
$$

avec $\mathrm{d}S_2 >= 0$.

---

On développe les intégrales et on obtient

$$
2 E \pi r^2 = \frac{Q_\text{int}}{\varepsilon_0}
$$

---

Ensuite on substitue Qint

$$
2 E \pi r^2 = \frac{\pi r^2 Q}{\varepsilon_0 S} \quad \text{avec } S \text{ la surface totale de la plaque.}
$$

---

On simplifie le tout pour isoler E

$$
E = \frac{Q}{2 \varepsilon_0 S}
$$

---

On obtient le résultat final suivant :

$$
\vec{E} = \frac{Q}{2 \varepsilon_0 S} \vec{e_y}
$$








### Trouvons l’angle incident final :

On considère le faisceau de particules arrivant avec un angle initial de $\theta$ avec $\vec{e_y}$ et une vitesse initiale $\vec{v}_0$.

On considère que les particules sont soumises au champ créé par la plaque à partir du point $(x_0, y_0)$.

D'après les équations d'une particule de charge $q$ et de masse $m$ dans un champ électrique :

$$
\begin{cases}
x(t) = \ v_0 t\cdot\vec{e_x}\ + x_0 \quad \text{(1)} \\ 
y(t) = \frac{q}{2m} E t^2\cdot\vec{e_y}\ + v_0 t\cdot\vec{e_y}\ + y_0 \quad \text{(2)}
\end{cases}
$$

---

A présent, calculons l'équation de la trajectoire.

À partir de (1), on exprime $t$ :

$$
t = \frac{x - x_0}{v_0 \cdot \vec{e_x}}
$$


Substituons cette valeur de $t$ dans (2) pour obtenir $y(x)$ :

$$
y(x) = \frac{q}{2m} \left( \frac{x - x_0}{v_0\cdot\vec{e_x}} \right)^2 E\cdot\vec{e_y} 
+ v_0\cdot\vec{e_y} \left( \frac{x - x_0}{v_0\cdot\vec{e_x}} \right) + y_0
$$


Simplifions :

$$
y(x) = \frac{q}{2m} \left( \frac{x - x_0}{v_0 \sin \theta} \right)^2 E 
- v_0 \cos \theta \left( \frac{x - x_0}{v_0 \sin \theta} \right) + y_0
$$

---

On dérive $y(x)$ et on obtient

$$
y'(x) = \frac{q E}{m (v_0 \sin \theta)^2} \left( x - x_0 \right) + \frac{1}{\tan \theta}
$$

---

A présent, on pose $x_s$ la solution de l’équation $y(x) = 0$.  
Alors l’équation de la tangente en $x_s$ dans le repère $(O_s, \vec{e_x}, \vec{e_y})$ où $O_s$ est le point $(x_s, 0)$, est :

$$
y = y'(x_s) x
$$


On en déduit : 

$$
\tan\left(\frac{\pi}{2} - \alpha\right) = y'(x_s)
$$

En utilisant la définition de la tangente :

$$
\frac{\sin\left(\frac{\pi}{2} - \alpha\right)}{\cos\left(\frac{\pi}{2} - \alpha\right)} = y'(x_s)
$$

D'après les identités trigonométriques :

$$
\frac{\cos\alpha}{\sin\alpha} = y'(x_s)
$$

En isolant $\alpha$ :

$$
\alpha = \tan^{-1}\left(\frac{1}{y'(x_s)}\right)
$$


### Trouvons (xs) afin de connaitre tous les termes dont $\alpha$ dépend :
On a que :
$$
y(x) = \frac{q}{2m} \left( \frac{x - x_0}{v_0 \sin \theta} \right)^2 E 
- v_0 \cos \theta \left( \frac{x - x_0}{v_0 \sin \theta} \right) + y_0
$$

Posons :
$$
X = \frac{x - x_0}{v_0 \sin \theta}
$$

On obtient :
$$
y(x) = \frac{q}{2m} \left(X\right)^2 E 
- v_0 \cos \theta \left(X\right) + y_0
$$


Calculons le discriminant $ \Delta$ : 
$$ \Delta = \left( v_0 \cos\theta \right)^2 - \frac{2 y_0 q E}{m} $$

On suppose $ \Delta$ positif car nos solutione ne peuvent pas etre imaginaires, et on obtient les racines suivantes :
$$
\begin{aligned}
X_1 &= \frac{v_0 \cos\theta - \sqrt{\left( v_0 \cos\theta \right)^2 - \frac{2 y_0 q E}{m}}}{\frac{q}{m} E} = \frac{x - x_0}{v_0 \sin \theta} \\[10pt]
X_2 &= \frac{v_0 \cos\theta + \sqrt{\left( v_0 \cos\theta \right)^2 - \frac{2 y_0 q E}{m}}}{\frac{q}{m} E} = \frac{x - x_0}{v_0 \sin \theta}
\end{aligned}
$$


Ce qui nous donne : 
$$\
xs = \left( \frac{v_0 \cos\theta - \sqrt{\left( v_0 \cos\theta \right)^2 - \frac{2 y_0 q E}{m}}}{\frac{q}{m} E} \right) v_0 \sin\theta + x_0
$$

On posera par la suite que x0 = 0 pour simplifier les expressions. On place donc le centre de notre repère à l'ordonnée de l'échantillon et à l'abscisse du point initial.

On utilise par la suite l'expression de xs pour calculer y'(xs) avant d'avoir l'expression finale de α.

# Calcul de l'incertitude sur xs

## Expression de xs
$$xs = \left( \frac{v_0 \cos\theta - \sqrt{\left( v_0 \cos\theta \right)^2 - \frac{2 y_0 q E}{m}}}{\frac{q}{m} E} \right) v_0 \sin\theta$$

## Calcul des dérivées partielles

Pour simplifier les calculs, posons:
- $A = v_0 \cos\theta$
- $B = \frac{2 y_0 q E}{m}$
- $C = \frac{q}{m} E$
- $D = \sqrt{A^2 - B}$

Alors:
$$xs = \frac{(A - D)}{C} \cdot v_0 \sin\theta$$

### 1. Dérivée par rapport à $v_0$

$v_0$ apparaît dans $A$ et le facteur multiplicatif $v_0 \sin\theta$.

$$\frac{\partial xs}{\partial v_0} = \frac{\partial}{\partial v_0}\left[\frac{(A - D)}{C} \cdot v_0 \sin\theta\right]$$

$$= \frac{\sin\theta}{C}(A - D) + \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial v_0}(A - D)$$

$$= \frac{\sin\theta}{C}(A - D) + \frac{v_0 \sin\theta}{C} \cdot \left(\cos\theta - \frac{A \cos\theta}{D}\right)$$

$$= \frac{\sin\theta}{C}(A - D) + \frac{v_0 \sin\theta \cos\theta}{C} \cdot \left(1 - \frac{A}{D}\right)$$

### 2. Dérivée par rapport à $\theta$

$\theta$ apparaît dans $A$ et dans $\sin\theta$.

$$\frac{\partial xs}{\partial \theta} = \frac{\partial}{\partial \theta}\left[\frac{(A - D)}{C} \cdot v_0 \sin\theta\right]$$

$$= \frac{v_0 \cos\theta}{C}(A - D) + \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial \theta}(A - D)$$

$$= \frac{v_0 \cos\theta}{C}(A - D) + \frac{v_0 \sin\theta}{C} \cdot \left(-v_0 \sin\theta - \frac{A \cdot (-v_0 \sin\theta)}{D}\right)$$

$$= \frac{v_0 \cos\theta}{C}(A - D) - \frac{v_0^2 \sin^2\theta}{C} \cdot \left(1 - \frac{A}{D}\right)$$

### 3. Dérivée par rapport à $y_0$

$y_0$ apparaît uniquement dans $B$.

$$\frac{\partial xs}{\partial y_0} = \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial y_0}(A - D)$$

$$= \frac{v_0 \sin\theta}{C} \cdot \left(0 - \frac{1}{2D} \cdot \frac{\partial B}{\partial y_0}\right)$$

$$= -\frac{v_0 \sin\theta}{C} \cdot \frac{1}{2D} \cdot \frac{2qE}{m}$$

$$= -\frac{v_0 \sin\theta \cdot qE}{C \cdot m \cdot D}$$

### 4. Dérivée par rapport à $q$

$q$ apparaît dans $B$ et $C$.

$$\frac{\partial xs}{\partial q} = \frac{\partial}{\partial q}\left[\frac{(A - D)}{C} \cdot v_0 \sin\theta\right]$$

$$= -\frac{(A - D)}{C^2} \cdot \frac{\partial C}{\partial q} \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial q}(A - D)$$

$$= -\frac{(A - D)}{C^2} \cdot \frac{E}{m} \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \left(0 - \frac{1}{2D} \cdot \frac{\partial B}{\partial q}\right)$$

$$= -\frac{(A - D) \cdot E \cdot v_0 \sin\theta}{C^2 \cdot m} - \frac{v_0 \sin\theta}{C} \cdot \frac{1}{2D} \cdot \frac{2y_0E}{m}$$

$$= -\frac{(A - D) \cdot E \cdot v_0 \sin\theta}{C^2 \cdot m} - \frac{v_0 \sin\theta \cdot y_0E}{C \cdot m \cdot D}$$

### 5. Dérivée par rapport à $m$

$m$ apparaît dans $B$ et $C$.

$$\frac{\partial xs}{\partial m} = \frac{\partial}{\partial m}\left[\frac{(A - D)}{C} \cdot v_0 \sin\theta\right]$$

$$= -\frac{(A - D)}{C^2} \cdot \frac{\partial C}{\partial m} \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial m}(A - D)$$

$$= -\frac{(A - D)}{C^2} \cdot \left(-\frac{qE}{m^2}\right) \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \left(0 - \frac{1}{2D} \cdot \frac{\partial B}{\partial m}\right)$$

$$= \frac{(A - D) \cdot qE \cdot v_0 \sin\theta}{C^2 \cdot m^2} - \frac{v_0 \sin\theta}{C} \cdot \frac{1}{2D} \cdot \left(-\frac{2y_0qE}{m^2}\right)$$

$$= \frac{(A - D) \cdot qE \cdot v_0 \sin\theta}{C^2 \cdot m^2} + \frac{v_0 \sin\theta \cdot y_0qE}{C \cdot m^2 \cdot D}$$

### 6. Dérivée par rapport à $E$

$E$ apparaît dans $B$ et $C$.

$$\frac{\partial xs}{\partial E} = \frac{\partial}{\partial E}\left[\frac{(A - D)}{C} \cdot v_0 \sin\theta\right]$$

$$= -\frac{(A - D)}{C^2} \cdot \frac{\partial C}{\partial E} \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \frac{\partial}{\partial E}(A - D)$$

$$= -\frac{(A - D)}{C^2} \cdot \frac{q}{m} \cdot v_0 \sin\theta + \frac{v_0 \sin\theta}{C} \cdot \left(0 - \frac{1}{2D} \cdot \frac{\partial B}{\partial E}\right)$$

$$= -\frac{(A - D) \cdot q \cdot v_0 \sin\theta}{C^2 \cdot m} - \frac{v_0 \sin\theta}{C} \cdot \frac{1}{2D} \cdot \frac{2y_0q}{m}$$

$$= -\frac{(A - D) \cdot q \cdot v_0 \sin\theta}{C^2 \cdot m} - \frac{v_0 \sin\theta \cdot y_0q}{C \cdot m \cdot D}$$

## Formule finale pour l'incertitude sur xs

L'incertitude sur $xs$ est donnée par la formule:

$$\Delta xs = \sqrt{\sum_i \left(\frac{\partial xs}{\partial x_i} \Delta x_i\right)^2}$$

En remplaçant les dérivées partielles calculées:

$$\Delta xs = \sqrt{\left(\frac{\partial xs}{\partial v_0} \Delta v_0\right)^2 + \left(\frac{\partial xs}{\partial \theta} \Delta \theta\right)^2 + \left(\frac{\partial xs}{\partial y_0} \Delta y_0\right)^2 + \left(\frac{\partial xs}{\partial q} \Delta q\right)^2 + \left(\frac{\partial xs}{\partial m} \Delta m\right)^2 + \left(\frac{\partial xs}{\partial E} \Delta E\right)^2}$$

Pour une application numérique, il faudrait connaître:
- Les valeurs de $v_0$, $\theta$, $y_0$, $q$, $m$, et $E$
- Les incertitudes associées $\Delta v_0$, $\Delta \theta$, $\Delta y_0$, $\Delta q$, $\Delta m$, et $\Delta E$
